```python
#droping the unwanted columns
data.drop(labels=["car_ID","symboling","CarName"],axis=1,inplace=True)
data.head()
```

```python
#checking unique categorical values in each columns
#These columns have categorical values we have to convert them into numarical values
for column in data:
    if data[column].dtypes=="object":
        print("----",column,"----"*10)
        print(data[column].unique())
        print(data[column].value_counts())
```

```python
#we have to convert only categorical values in numrical values using 
from sklearn.preprocessing import LabelEncoder
```

```python
label_en = LabelEncoder()
categorical_columns=["fueltype","aspiration","doornumber","carbody","drivewheel","enginelocation","enginetype","cylindernumber","fuelsystem"]
#data[categorical_columns] = label_en.fit_transform(data[categorical_columns])
coverted_columns=data[categorical_columns].apply(label_en.fit_transform)
coverted_columns
```

```python
labels=['fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel','enginelocation', 'enginetype', 'cylindernumber', 'fuelsystem']
data.drop(labels=labels,axis=1,inplace=True)
data
```

```python
dataset=pd.concat([coverted_columns,data],axis=1)
dataset.head()
```

```python
#spliting x and y datas
dataset1=dataset.copy()
```

```python
x = dataset.drop('price',axis = 1)
y = dataset['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_prediction=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_prediction)
```
