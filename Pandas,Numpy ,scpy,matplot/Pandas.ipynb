```python
import pandas as pd
import numpy as np
```

```python
#creating series using a array without index
arr=np.array([1,2,3,4,5,6,7])
series_1=pd.Series(arr)
print(series_1)
print(type(series_1))
```

```python
##creating series using a array with index
arr=np.array(["a","b","c","d","e"])
s_1=pd.Series(arr,index=[11,12,13,14,15])
s_1
```

```python
#Dictionary
#here keys will be our default indexes
my_dict={
    "name":["sagar","brahma","abhi","harish"],
    "scores":[65,451,155,21]
}
#passing my dictinary into series
my_series=pd.Series(my_dict)
my_series
```

```python
name=["sagar","brahma","abhi","harish"]
scores=[65,451,155,21]
s=pd.Series(scores,index=name)
print(s)
```

```python
#Acessing items in series
s=pd.Series(["apple","mango","banana","Grapes","kivi","pinaple"])
print(s)
#Accessing with index
print(s[1])
print(s[3])
print(s[1:])
```

```python
name=["sagar","brahma","abhi","harish"]
scores=[65,451,155,21]
s=pd.Series(scores,index=name)
print(s)
print("____________________________________")
print(s.index)#to get the index
print(s.values)#to get the values
print("_______________")
#Aceesing items
print("scores of sagar=",s["sagar"])
print("scores of abhi=",s["abhi"])
```

```python
#creating a dataframe by dictionary
names=["sagar","abhi","brahma","basava","ajaya"]
scores=[120,165,151,123,154]
dict_1={
    "names":names,
    "scores":scores
}
print("Dictionary =",dict_1)
print("_______________________")
dataFrame=pd.DataFrame(dict_1)
dataFrame
```

```python
#creating a dataframe and adding index and colomn names
x=["Tata","ford","KIa","tesla","Maruthi","mahendra"]
y=[14,55,11,45,55,11]

dicti={
    0:x,
    1:y
}
df=pd.DataFrame(dicti, index=["c1","c2","c3","c4","c5","c6"],columns=["company","sales"])
df["company"]=x
df["sales"]=y
#adding extra data to a data frame
year=["2001","2022","2015","2016","2019","2020"]
empty=np.arange(0,6)
df["Year"]=year
df["empty"]=empty
df
#Deleting empty column #or use pop() method to remove
del df["empty"]
df
```

```python
#some important properties and functions of pandas
#arrays
random=np.random.randint(low=0, high=100, size=20)
name=np.random.choice(["a","b","c","d"],size=20)
choice=np.random.choice([10,11,12,13,14,15],size=20)
print(random)
print(name)
print(choice)
```

```python
#creating empty dataframe and adding above arrays to 
data=pd.DataFrame()
data["Random"]=random
data["name"]=name
data["choice"]=choice
data
```

```python
#shape of dataframe
print("shape of dataframe",data.shape)
#column names
print("column names",data.columns)
#column values
print("column values",data.values)
```
