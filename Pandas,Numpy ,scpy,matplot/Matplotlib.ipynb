```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating array's for ploting
x=np.arange(0,25)
y=np.arange(10,35)

#useing scatter for scatter plot and put the x and y values
plt.scatter(x,y,c="b",marker=".")# c=color,marker=scatter style
#providing title and x and y label for graph
plt.title("Scatter plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
#line plot
plt.plot(x,y,c="red")
plt.title("Line plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
#pi in numpy
np.pi
```

```python
#x1 are the values from 0 to pi(3.1415..) values in the step of 0.1  
x1=np.arange(0,np.pi,0.1)
#Y1 are values of sine of each x1 values
y1=np.sin(x1)
print("values from 0 to 3.14 =",x1)
print("")
print("sin values from 0 to 3.14 =",y1)
```

```python
#Sin wave from 0 to 2 multiply of pi
x1=np.arange(0,2*np.pi,0.1)
y1=np.sin(x1)

plt.plot(x1,y1,"blue")
plt.title("Sin wave")
plt.xlabel("X1 values")
plt.ylabel("y1 values")
plt.show()
```

```python
x1=np.arange(0,2*np.pi,0.1)
y1=np.cos(x1)

plt.plot(x1,y1,"red")
plt.title("Sin wave")
plt.xlabel("X1 values")
plt.ylabel("y1 values")
plt.show()
```

```python
x1=np.arange(0,2*np.pi,0.1)
y1=np.tan(x1)

plt.plot(x1,y1,"blue")
plt.title("tan wave")
plt.xlabel("X1 values")
plt.ylabel("y1 values")
plt.show()
```

```python
#bar graph
#ex:The following data represents the sale of refrigerator sets in a showroom in first 6 months of the year.
months=np.array(["jan","feb","march","april","may","june"])
No_of_Refrigerators_Sold=np.array([26,65,48,29,65,63])
#ploting
plt.bar(months,No_of_Refrigerators_Sold)
plt.title("Months vs sales")
plt.xlabel("months")
plt.ylabel("Sold refrigerators")
plt.show()
```

```python
x = np.array([22,87,5,43,56,8,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(x)
plt.show()
```

```python
x = np.array([22,87,5,43,56,8,73,65,55,54,11,20,51,5,79,31,27]) 
plt.hist(x,bins=10)
plt.xlabel('X values')
plt.ylabel('No. of values')
plt.xticks(np.arange(0,100,5))
plt.yticks(np.arange(0,5))
plt.show()
```

```python
#box plot used for removing outlier from data
data=pd.Series([10,20,54,89,62,14,36,65,200,-3])
plt.boxplot(data)
plt.show()
```

```python
#pie chart
data=[65,32,85,20]
label=['apple','mango','orange','plum']
colors = ['r', 'g', 'b', 'y']
explode=(0,0,0.1,0.23)

plt.pie(data,explode=explode,labels=label,colors=colors,autopct='%1.1f%%',shadow=True)
plt.title('Fruits vs sales')
plt.show()
```

```python
#creating subplots
x1=np.arange(10)
y1=x1**2

plt.figure(figsize=(10,10))
#first subplot
plt.subplot(2,2,1)
plt.plot(x1,y1,'bo')# "bo"= using blue circle markers
plt.title("fig 1")

#second subplot
plt.subplot(2,2,2)
plt.plot(x1,y1,'go')
plt.title("fig 2")

#third subplot
plt.subplot(2,2,3)
plt.plot(x1,y1,'D')# d-diamond marker
plt.title("fig 3")

#fourth subplot
plt.subplot(2,2,4)
plt.plot(x1,y1,'g*--',)#green color with dashed line
plt.title("fig 4")
plt.show()
```
