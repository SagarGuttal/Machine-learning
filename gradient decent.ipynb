## f(x)=x**2+x+1

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x_values=np.linspace(start=-3,stop=3,num=300)
x_values
```

### Cost function

```python
def cost_function(x):
    return x**2+x+1
```

### Derivative of cost function

```python
def derivative_of_cost_function(x):
    return 2*x+1
```

```python
# x_values to cost function
cost_function(x_values)[:10]
```

```python
# x_values to derivative of cost function
derivative_of_cost_function(x_values)[:10]
```

### Plot the cost function

```python
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
plt.title("Simple cost function",fontsize=18)
sns.lineplot(x=x_values,y=cost_function(x_values),color="blue")
plt.grid(visible=True)
plt.xlabel( 'X values', fontsize = 18)
plt.ylabel ('df(x)',fontsize = 18 )

plt.subplot(1,2,2)
plt.title("Derivative of Simple cost function",fontsize=18)
sns.lineplot(x=x_values,y=derivative_of_cost_function(x_values),color="blue")
plt.grid(visible=True)
plt.xlabel( 'X values', fontsize = 18)
plt.ylabel ('df(x)',fontsize = 18 )
plt.show()
```

```python
new_x=3
previous_x=0
Alpha=0.1
x_list=[]
slope_list=[]

for weight in range(100):
    previous_x=new_x
    new_x=previous_x-Alpha*derivative_of_cost_function(x_values)
    x_list.append(new_x)
    slope_list.append(derivative_of_cost_function(new_x))
print(new_x)
```
